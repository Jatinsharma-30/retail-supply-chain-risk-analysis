```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../data/Raw Data/retail_store_inventory.csv')
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.info()
```

```python
df.duplicated().sum()
```

# No duplicate records found

```python
df['Year'] = df['Date'].dt.year
```

```python
df['Month'] = df['Date'].dt.month
```

```python
df['Month_Name'] = df['Date'].dt.month_name()
```

```python
df['Quarter'] = df['Date'].dt.quarter
```

```python
df['Forecast_Error'] = df['Units Sold'] - df['Demand Forecast']
```

```python
df['Absolute_Forecast_Error'] = (
    df['Units Sold']
    - df['Demand Forecast']
).abs()
```

```python
df['Inventory_Value'] = (
    df['Inventory Level']
    * df['Price']
)
```

```python
df['Discount_Flag'] = np.where(
    df['Discount'] > 0,
    'Discounted',
    'Non-Discounted'
)
```

```python
df['Stock_Coverage'] = (
    df['Inventory Level']
    / df['Demand Forecast']
)
```

```python
df['Inventory_Risk'] = np.where(
    df['Inventory Level'] < df['Demand Forecast'],
    'High Risk',
    'Low Risk'
)
```

```python
df['Price_Difference'] = (
    df['Price']
    - df['Competitor Pricing']
)
```

```python
df.to_csv(
    '../data/cleaned/retail_inventory_cleaned.csv',
    index=False
)
```
